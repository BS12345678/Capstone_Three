```python
#Importing necessary modules
import pandas as pd
pd.options.mode.chained_assignment = None #so that when I slice the data, it doesn't give a warning 
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import pickle
```

```python
#importing data from preprocessing
df6 = pd.read_csv('...Data\cc_2.csv')
```

```python
df6.columns
```

```python
# Adding numeric columns and dummy columns side by side
neworder = ['Unnamed: 0', 'CustomerID', 'Tenure Months', '#ofServices', 'Churn Value', 'CLTV',
       'Average Monthly Charges', 'City_churn_rate', 'Senior Citizen_Yes',
       'Partner_Yes', 'Dependents_Yes', 'Phone Service_Yes',
       'Multiple Lines_Yes', 'Internet Service_DSL',
       'Internet Service_Fiber optic', 'Online Security_Yes',
       'Online Backup_Yes', 'Device Protection_Yes', 'Tech Support_Yes',
       'Streaming TV_Yes', 'Streaming Movies_Yes', 'Contract_Month-to-month',
       'Contract_One year', 'Paperless Billing_Yes',
       'Payment Method_Bank transfer (automatic)',
       'Payment Method_Credit card (automatic)',
       'Payment Method_Electronic check']
df6=df6.reindex(columns=neworder)
```

```python
df6.info()
```

**Separating features and target**

```python
#dropping unwanted columns
df6 = df6.drop(columns=['Unnamed: 0','CustomerID','CLTV','City_churn_rate'], axis=1)
```

```python
X = df6.drop('Churn Value', axis=1)
y = df6['Churn Value']
```

**Splitting X and y into train and test samples**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
X_train.head()
```

**Standardizing numeric features**

```python
scaler = StandardScaler()
X_train.iloc[:,0:3] = scaler.fit_transform(X_train.iloc[:,0:3])
X_test.iloc[:,0:3] = scaler.transform(X_test.iloc[:,0:3])
```

```python
X_train.head(2)
```

```python
X_test.head(2)
```

# **Modeling**

<font color='blue'>1. DUMMY REGRESSOR (using mode)</font>

```python
# Initially can use a dummyregressor
#dumb_reg is an array of zeros as it's the most occuring labels in y_train
# here I used the mode value
dumb_reg = stats.mode(y_train)
dumb_reg
```

```python
#Calculating accuracy (accuracy = correct predictions/total number of predictions)
# number of zeros in y_test = y_test[y_test==0].count()
#total number of predictions = y_test.shape[0]
dumb_reg_accuracy = (y_test[y_test==0].count()/y_test.shape[0])*100
```

```python
print('dumb_reg_accuracy is {0:.2f}'.format(dumb_reg_accuracy))
```

<font color='blue'>2. Logistic Regression</font>

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
y_pred_prob = lr.predict_proba(X_train)[:,1]
y_pred_prob1 = lr.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_train, y_pred_prob)
fpr1, tpr1, thresholds1 = roc_curve(y_test, y_pred_prob1)

print('Logistic regression_train_accuracy is {0:.3f}'.format(accuracy_score(lr.predict(X_train), y_train)))
print('Logistic regression_test_accuracy is {0:.3f}'.format(accuracy_score(lr.predict(X_test), y_test)))
print('Logistic regression_train_auc is {0:.3f}'.format(auc(fpr, tpr)))
print('Logistic regression_test_auc is {0:.3f}'.format(auc(fpr1, tpr1)))
```

```python
# To find the coeffiencts of logistic regression, I merged the coefficients with the column names
coefficients = pd.concat([pd.DataFrame(X_train.columns,columns=['features']),pd.DataFrame(np.transpose(lr.coef_), columns=['coeff'])], axis = 1)
```

```python
# Coefficients' values sorted in descending order by their absolute values.
coefficients.reindex(coefficients.coeff.abs().sort_values(ascending=False).index)
```

<font color='blue'>3.Logistic Regression with GridSearchCV</font>

```python
from sklearn.model_selection import GridSearchCV
lr = LogisticRegression(solver = 'liblinear', penalty='l1', max_iter=100)
Cs = [0.001, 0.01, 0.05, 0.1, 1]
parameters = {'C': Cs}
cv = GridSearchCV(lr, parameters, cv=20)
cv.fit(X_train, y_train)
print(cv.best_params_)
```

```python
y_pred_proba = cv.predict_proba(X_train)[:,1]
y_pred_proba1 = cv.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_train, y_pred_proba)
fpr_lr, tpr_lr, thresholds = roc_curve(y_test, y_pred_proba1)
precision_lr, recall_lr, thresholds = precision_recall_curve(y_test, y_pred_proba1)
```

```python
print('Logistic regression with gridsearchCV train accuracy is {0:.3f}'.format(accuracy_score(cv.predict(X_train), y_train)))
print('Logistic regression with gridsearchCV test accuracy is {0:.3f}'.format(accuracy_score(cv.predict(X_test), y_test)))
print('Logistic regression with gridsearchCV train auc is {0:.3f}'.format(auc(fpr, tpr)))
print('Logistic regression with gridsearchCV test auc is {0:.3f}'.format(auc(fpr_lr, tpr_lr)))
```

```python
y_pred = cv.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

<font color='blue'>5.XGBoost with hyperparameters tuned</font>

```python
max_depth = [3,4,5]
learning_rate = [0.01, 0.05, 0.10]
subsample = [0.8]
colsample_bytree = [0.5,0.75]
n_estimators = [100]
min_child_weight = [1,2]
param_grid = {
    "max_depth": max_depth,
    "learning_rate": learning_rate,
    "subsample": subsample,
    "colsample_bytree": colsample_bytree,
    "n_estimators": n_estimators,
    "min_child_weight" : min_child_weight
}
```

```python
y_pred_proba = grid_cv.predict_proba(X_train)[:,1]
y_pred_proba1 = grid_cv.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_train, y_pred_proba)
fpr_xgb, tpr_xgb, thresholds1 = roc_curve(y_test, y_pred_proba1)
precision_xgb, recall_xgb, thresholds = precision_recall_curve(y_test, y_pred_proba1)
```

**How does this model matter to the business?**\
<font color='blue'>Adjusting Decision Boundary to optimize the model for better performance</font>

Since the test_auc and test_accuracy were higher for XGBoost(with GridsearchCV) than other models, it was decided as the final model. However the recall for label 1 (or churners) was lower (0.57) meaning the false negatives were higher, because 
recall = tp/(tp +fn). That is, coniderable amount of actual churners were not detected.
In a Telecom industry, not being able to find a churner will affect the business adversely. That is, sensitivity/tpr/recall matters a lot. Therefore to improve detecting higher number of true positives, I decided to reduce the decision boundary (usually it is 0.5) to the range of 0.3 - 0.4. 


```python
#calculating predicted probability
y_pred_proba1 = grid_cv.predict_proba(X_test)[:,1]

# Giving predicted labels (0 or 1) based on decision boundary 0.3-0.4
y_pred_n = list()
for i in y_pred_proba1:
    if i >= 0.33:
        y_pred_n.append(1)
    else:
        y_pred_n.append(0)
print(confusion_matrix(y_test, y_pred_n))
print(classification_report(y_test, y_pred_n))
```

After trying different values of decision boundaries, it was decided that when decision boundary is 0.33 (as opposed to 0.5),
the model identifies around 283 churners out of 362 total actual churners with a recall of 0.78 (earlier it was 0.57).
While there's an improvement in the recall, the precision which equals to tp/(tp+fp) has gone down as we over predict 
the number of positives (i.e. that is the false positives). However, in this situation, not detecting an acutal churner is worse than identifying a non-churner as a churner. Also one should be careful of false positives, as it might lead to unnecessary costs targeted at customers who may not churn. 
Hence, it's important that we optimize the level of decision boundary so that the recall doesn't cost much of precision.
